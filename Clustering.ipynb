# Hierarchical - Cluster

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import linkage 
import scipy.cluster.hierarchy as sch
```

```python
df=pd.read_csv("crime_data.csv")
```

```python
df
```

```python
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
df_norm = norm_func(df.iloc[:,1:])
```

```python
df_norm
```

```python
dendrogram=sch.dendrogram(sch.linkage(df_norm,method='complete'))
```

```python
from sklearn.cluster import AgglomerativeClustering
hc=AgglomerativeClustering(n_clusters=3,linkage='complete',affinity="euclidean").fit(df_norm)
cluster_labels=pd.Series(hc.labels_)
cluster_labels
df['clust']=cluster_labels
df
```

```python
df.iloc[:,1:].groupby(df.clust).mean()
```

```python
df.sort_values("clust")
```

```python
df.to_csv('hc.csv')
```

# K - Mean

```python
df=pd.read_csv("crime_data.csv")
df
```

```python
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
df_norm = norm_func(df.iloc[:,1:])
df_norm
```

```python
from sklearn. cluster import KMeans
WCSS = []
for i in range(1, 11):
    km= KMeans(n_clusters=i)
    km.fit(df_norm)
    WCSS.append(km.inertia_)
plt.plot(range(1, 11), WCSS)
plt.title('The Elbow Method')
plt.ylabel('WCSS')
plt.xlabel('Number of Clusters')
plt.show()
```

```python
km=KMeans(n_clusters=3)
y_KMeans=km.fit_predict(df_norm)
```

```python
y_KMeans
```

```python
md=pd.series(y_KMeans)
df['clust']=md
df
```

```python
df.iloc[:,1:7].groupby( df.clust).mean()
```

```python
import sklearn
sklearn.metrics.silhouette_score(df_norm, y_KMeans)
```

```python
df.plot(x="Murder",y="Rape",c=km.labels_,kind="scatter",s=50 ,cmap=plt.cm.coolwarm)
plt.title('Cluster using KMeans')
```

# DB SCAN -CLUSTERING

```python
df=pd.read_csv("crime_data.csv")
```

```python
df
```

```python
df.info()
```

```python
array=df.iloc[:,1:]
array
```

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
stsclar = StandardScaler().fit(array)
x = stsclar.transform(array)
x 
```

```python
dbscan = DBSCAN(eps=1, min_samples=7)
dbscan.fit(x)
```

```python
dbscan.labels_
```

```python
c3=pd.DataFrame(dbscan.labels_,columns=['cluster'])
c3
```

```python
df2=pd.concat([df,c3],axis=1)
df2
```

```python
import sklearn
sklearn.metrics.silhouette_score(x, df2)
```

```python
df2.plot(x="Murder",y="Rape",c=dbscan.labels_,kind="scatter",s=100,cmap=plt.cm.copper_r)
plt.title('Cluster using DBScan')
```
